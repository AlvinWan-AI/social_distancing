```python
import pandas as pd
```

```python
df = pd.read_csv("data/2020_CA_Region_Mobility_Report.csv")
```

```python
df.head()
```

```python
ontario_mobility = df[(df.sub_region_1 == "Ontario") & (df.sub_region_2.isna())]
```

```python
ontario_mobility
```

```python
ontario_mobility.drop(['country_region_code', 'iso_3166_2_code', 'place_id'], axis=1, inplace=True)
```

```python
ontario_mobility
```

```python
ontario_mobility.isna().sum()
```

```python
ontario_mobility.drop(['metro_area', 'census_fips_code'], axis=1, inplace=True)
```

```python
ontario_mobility[ontario_mobility.isna().any(axis=1)]
```

```python
ontario_mobility
```

```python
ontario_mobility['parks_percent_change_from_baseline'] = ontario_mobility.groupby(['sub_region_2'], sort=False)['parks_percent_change_from_baseline'].apply(lambda x: x.ffill().bfill())
```

```python
ontario_mobility['transit_stations_percent_change_from_baseline'] = ontario_mobility.groupby(['sub_region_2'], sort=False)['transit_stations_percent_change_from_baseline'].apply(lambda x: x.ffill().bfill())
```

```python
ontario_mobility["sd_score"] = (-1) *(ontario_mobility["retail_and_recreation_percent_change_from_baseline"] + ontario_mobility["grocery_and_pharmacy_percent_change_from_baseline"] + ontario_mobility["parks_percent_change_from_baseline"] + ontario_mobility["transit_stations_percent_change_from_baseline"] + ontario_mobility["workplaces_percent_change_from_baseline"]) + ontario_mobility["residential_percent_change_from_baseline"] 
```

```python
ontario_mobility
```

```python
min_score = ontario_mobility.sd_score.min()
max_score = ontario_mobility.sd_score.max()
```

```python
min_score
```

```python
max_score
```

```python
scorelist = ontario_mobility.sd_score.tolist()
```

```python
final_max = 100
final_min = 0
```

```python
ontario_mobility["standarized_sd_score"] = ((ontario_mobility.sd_score - min_score) / (max_score - min_score)) * (final_max - final_min) + final_min
```

```python
ontario_mobility
```

```python
ontario_mobility.to_csv('data/ON_mobility_state_only.csv', index=False)  
```
